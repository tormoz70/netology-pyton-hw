### 3. Рассмотрим какие жанры наиболее популярны у создателей фильмов 

_Круговая диаграмма удобна для отображения доли вклада каждой анализируемой категории в общее множество. Исходя из графика можно сразу отметить что наиболее популярны у зрителей такие жанры как "боевики", "приключения", и "триллеры"._

```python
df2 = df[['genres', 'budget', 'popularity', 'revenue']]
df2.head()
```

```python
import json
ggg = {}
added = 0
for row in df2.iterrows():
    gnrs = json.loads(row[1].genres)
    for r in gnrs:
        if (r['id'] not in ggg):
            added += 1
            ggg[r['id']] = {'name': r['name'], \
                            'count': 1,
                            'budget': row[1].budget//1000000, \
                            'revenue': row[1].revenue//1000000, \
                            'popularitySum': row[1].popularity
                           }
        else:
            ggg[r['id']]['count'] += 1
            ggg[r['id']]['budget'] += row[1].budget//1000000
            ggg[r['id']]['revenue'] += row[1].revenue//1000000
            ggg[r['id']]['popularitySum'] += row[1].popularity

gggData = []
for key in ggg:
    gggData.append({'id':key, \
                    'name': ggg[key]['name'], \
                    'count': ggg[key]['count'], \
                    'budget': ggg[key]['budget'], \
                    'revenue': ggg[key]['revenue'], \
                    'popularitySum': ggg[key]['popularitySum']})
genres1 = pd.DataFrame(gggData)
genres1['popularity'] = genres1.apply(lambda x: x['popularitySum']/x['count'], axis=1)
genres1 = genres1.sort_values('popularity', ascending=False).head(10)
genres1.index = genres1['name']
genres1.head()
#added
```

### 4. Сравнение доходности нескольких кинокомпаний за весь период

_Это самый один распространенных видов визуализации "точечный график". На нем удобно наблюдать быстро меняющиеся процессы во времени, например, или распределения величин. Его принято использовать там где нужно отобразить данные с высокой точностью. Здесь мы видим как менялась доходность нескольких компаний во времени и можем сравнить их успешность на любой момент времени в заданном интервале. Так, на протяжении всего периода наиболее доходной можно назвать компанию "Columbia Pictures", но в 90-х "Walt Disney Pictures" явно всех обскакал, наверное, это связано с внедрением компьютерной графики в мультипликации._

```python
#'Walt Disney Pictures'
#df3 = df[df['production_companies'].str.contains('Walt Disney')][['release_date','production_companies','budget','revenue']]
df3 = df[['release_date','production_companies','budget','revenue']]
df3['year'] = df3['release_date'].apply(lambda x: str(x)[:4])
#df3['month'] = df3['release_date'].apply(lambda x: str(x)[5:7])
df3['margin_WD'] = df3.apply(lambda x: x['revenue']-x['budget'] if 'Walt Disney Pictures' in x['production_companies'] else 0, axis=1)
df3['margin_CP'] = df3.apply(lambda x: x['revenue']-x['budget'] if 'Columbia Pictures' in x['production_companies'] else 0, axis=1)
df3['margin_LP'] = df3.apply(lambda x: x['revenue']-x['budget'] if 'Legendary Pictures' in x['production_companies'] else 0, axis=1)
df3['margin_IFP'] = df3.apply(lambda x: x['revenue']-x['budget'] if 'Ingenious Film Partners' in x['production_companies'] else 0, axis=1)
df3 = df3.groupby(['year']).agg({'margin_WD':'sum','margin_CP':'sum', 'margin_LP':'sum', 'margin_IFP':'sum'}).reset_index().sort_values(['year'], ascending=[False])
df3.head()
```

```python
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
trace0 = go.Scatter(
    x=df3.year,
    y=df3.margin_WD,
    name='Walt Disney Pictures'
    #fill='tozeroy'
)

trace1 = go.Scatter(
    x=df3.year,
    y=df3.margin_CP,
    name='Columbia Pictures'
    #fill='tozeroy'
)

trace2 = go.Scatter(
    x=df3.year,
    y=df3.margin_LP,
    name='Legendary Pictures'
    #fill='tozeroy'
)

trace3 = go.Scatter(
    x=df3.year,
    y=df3.margin_IFP,
    name='Ingenious Film Partners'
    #fill='tozeroy'
)

data = [trace0, trace1, trace2, trace3]
layout = go.Layout(
    title='Margin of film companies',
    xaxis=dict(
        title='Год',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Прибыль, $',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

fig = go.Figure(data=data, layout=layout)

iplot(fig, show_link=False)
```

### 5. Как влияет сезон выхода фильма на его прибыльность

_Здесь я попытался проанализировать как на успешность фильма влиеяет время года, когда фильм выходит в прокат. Для отображения использовал классическую гистограмму. Это удобный инструмент для быстрого отображения некоторого распределения величины. Из график можно предположить, что выгоднее выпускать в прокат фильмы в конце весны-начале лета или в конце года. Хотя, может быть просто тенденция основана, на том, что просто компании чаще выпускают фильмы в эти периоды. Это уже можно сказать при более детальном исследовании._

```python
def get_season(release_date, mode=0):
    """Возвращмет время года"""
    if str(release_date)[5:7] in ['03','04','05']:
        return '01-Весна' if mode==0 else '01'
    elif str(release_date)[5:7] in ['06','07','08']:
        return '02-Лето' if mode==0 else '02'
    elif str(release_date)[5:7] in ['09','10','11']:
        return '03-Осень' if mode==0 else '03'
    elif str(release_date)[5:7] in ['12','01','02']:
        return '04-Зима' if mode==0 else '04'
    
```

```python
df4 = df[['release_date','budget','revenue']]
#df4['season'] = df4['release_date'].apply(get_season)
df4['month'] = df4['release_date'].apply(lambda x: str(x)[5:7])
df4['margin'] = df4.apply(lambda x: x['revenue']-x['budget'], axis=1)
df4 = df4.groupby(['month']).agg({'margin':'sum'}).reset_index().sort_values('month', ascending=True)
df4.head()
```

```python
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

```python

trace0 = go.Bar(
    x=df4.month,
    y=df4.margin
)

layout = go.Layout(
    title='Зависимость прибыльности фильма от времени года выхода',
    xaxis=dict(
        title='Месяц выхода в прокат',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Прибыль, $',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

fig = go.Figure(data=[trace0], layout=layout)

iplot(fig, show_link=False)
```
