## 1. Классическая теория вероятностей

#### Вопрос 1.

(a) - Для того что бы ладьи не били друг друга они должны находиться каждая на одной горизонтальной полосе, далее все определяется количеством перестановок по вертикали - n! (назовем их "удачными расстановками"), всего возможных комбинаций расстановок ладей на доске - $C_{n^{2}}^{n}$. Следовательно вероятность того, что они не бьют друг друга, это отношение "удачных расстоновок" к общему кол-ву расстановок - p = $\frac{n!}{C_{n^{2}}^{n}}$

(b) - Рассуждения такие же как в предыдущем вопросе, тольно из всех возможных "удачных расстановок" необходимо исключить одну - где все фигуры стоят на главной диагонали, следовательно p = $\frac{n!-1}{C_{n^{2}}^{n}}$

#### Вопрос 2. (условие не очень понятно написано, опишу как я понял)

(a) - брошено 3 кости, известно что на одной выпало 6, какова вероятность что на остальных двух тоже выпало по шестерке. Ответ: p = ${\frac{1}{6}}*{\frac{1}{6}}=\frac{1}{36}$

(a) - брошено 3 кости, известно что на двух выпало одинаковое число, какова вероятность что на всех по шестерке. Ответ: вероятность того, что на двух костях, на которых выпало одинаковое число и это чисто 6 -  $p_{1}$ = $\frac{1}{6}$, вероятность что на оставшейся кости также выпало 6 - $p_{2}$ = $\frac{1}{6}$, общая вероятность p = $p_{1}*p_{2}=\frac{1}{36}$
