(a) - брошено 3 кости, известно что на одной выпало 6, какова вероятность что на остальных двух тоже выпало по 
шестерке. 

P(A|B) = $\frac{P({A}\cap{B})}{P(B)}$, где A-вероятность что выпали 3 шестерки, B-вероятность что выпала одна шестерка. $P({A}\cap{B})$ = P(A), т.к. множество событий B включает в себя множество событий A, тогда объединение есть событие B, т.е. P(A|B) = $\frac{P(A)}{P(B)}$

вероятность того что выпали все шестерки P(A) = $\frac{1}{6^3}$, вероятность что выпала хотябы одна шестерка P(B) = 1-P($\overline{B}$), где P($\overline{B}) = (\frac{5}{6})^3$

Ответ: p = $\frac{({\frac{1}{6}})^3}{1-(\frac{5}{6})^3}$ = $\frac{1}{6^3-5^3}$
