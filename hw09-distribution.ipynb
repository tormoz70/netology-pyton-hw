### Распределение Пуассона

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from scipy.stats import poisson
```

```python
mu_valeus = [1, 5, 15]
linestyles = ['-', '--', ':']
colors = ['b', 'r', 'g']

fig, ax = plt.subplots(figsize=(5, 3.75))

for mu, ls, cl in zip(mu_valeus, linestyles, colors):
    dist = poisson(mu)
    x = np.arange(-1, 200)

    plt.plot(x, dist.pmf(x), color=cl,
             label=r'$\mu=%i$' % mu, linestyle=ls)

plt.xlim(-0.5, 30)
plt.ylim(0, 0.4)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\mu)$')
plt.title('Poisson Distribution')

plt.legend()
plt.show()
```

```python
mu_valeus = [1, 5, 15]
linestyles = ['-', '--', ':']
colors = ['b', 'r', 'g']

fig, ax = plt.subplots(figsize=(5, 3.75))

for mu, ls, cl in zip(mu_valeus, linestyles, colors):
    dist = poisson(mu)
    x = np.arange(-1, 200)

    plt.plot(x, dist.cdf(x), color=cl,
             label=r'$\mu=%i$' % mu, linestyle=ls)

plt.xlim(-2.5, 30)
plt.ylim(0, 1.1)

plt.xlabel('$x$')
plt.ylabel(r'$F(x)$')
plt.title('Poisson Distribution')

plt.legend()
plt.show()
```

### Экспоненциальное распределение

```python
from scipy.stats import expon
```

```python
mu_valeus = [1, 1.5, 3]
linestyles = ['-', '--', ':']
colors = ['b', 'r', 'g']

fig, ax = plt.subplots(figsize=(5, 3.75))

for mu, ls, cl in zip(mu_valeus, linestyles, colors):
    dist = expon(0, mu)
    x = np.arange(0, 15, 0.1)

    plt.plot(x, dist.pdf(x), color=cl,
             label=r'$\mu=%.1f$' % mu, linestyle=ls)

plt.xlim(0, 15)
plt.ylim(0, 1)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\mu)$')
plt.title('Exponential Distribution')

plt.legend()
plt.show()
```

```python
mu_valeus = [1, 1.5, 3]
linestyles = ['-', '--', ':']
colors = ['b', 'r', 'g']

fig, ax = plt.subplots(figsize=(5, 3.75))

for mu, ls, cl in zip(mu_valeus, linestyles, colors):
    dist = expon(0, mu)
    x = np.arange(0, 15, 0.1)

    plt.plot(x, dist.cdf(x), color=cl,
             label=r'$\mu=%.1f$' % mu, linestyle=ls)

plt.xlim(0, 15)
plt.ylim(0, 1)

plt.xlabel('$x$')
plt.ylabel(r'$F(x)$')
plt.title('Exponential Distribution')

plt.legend()
plt.show()
```
